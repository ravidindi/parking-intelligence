```python
text_columns = [
    "event_type",
    "event_cause",
    "description",
    "reason_breakdown"
]


df["combined_text"] = (
    df[text_columns]
    .fillna("")
    .astype(str)
    .agg(" ".join, axis=1)
    .str.lower()
)
```

```python
df["combined_text"].sample(10)
```

```python
parking_keywords = [

    "parking",
    "parked",
    "park",

    "illegal parking",
    "wrong parking",

    "stopped vehicle",
    "standing vehicle",

    "abandoned",

    "blocking",
    "obstruction",
    "lane block"
]
```

```python
df["is_parking_event"] = (
    df["combined_text"]
    .apply(
        lambda text:
        int(
            any(
                keyword in text
                for keyword in parking_keywords
            )
        )
    )
)
```

```python
df["is_parking_event"].value_counts()
```

```python
df[
    df["is_parking_event"] == 1
][
[
"event_type",
"event_cause",
"description",
"reason_breakdown"
]
].sample(20)
```

```python
parking_df = df[
    df["is_parking_event"] == 1
]


hotspots = (
    parking_df
    .groupby(
        [
        "junction",
        "zone"
        ]
    )
    .size()
    .reset_index(
        name="parking_events"
    )
    .sort_values(
        "parking_events",
        ascending=False
    )
)


hotspots.head(10)
```

```python
df["event_cause"].value_counts().head(50)
df["reason_breakdown"].value_counts().head(50)
df[
[
"event_cause",
"description",
"reason_breakdown"
]
].sample(30)
keywords = [
    "park",
    "parking",
    "vehicle",
    "stop",
    "halt",
    "obstruct",
    "block",
    "truck",
    "car",
    "bus"
]


for word in keywords:
    count = (
        df["combined_text"]
        .str.contains(
            word,
            na=False
        )
        .sum()
    )

    print(
        word,
        count
    )
    df["event_cause"].unique()
    df["reason_breakdown"].unique()
```

```python
obstruction_causes = [
    "vehicle_breakdown"
]


df["is_obstruction_event"] = (
    df["event_cause"]
    .isin(obstruction_causes)
    .astype(int)
)
df["is_obstruction_event"].value_counts()
```

```python
df["start_datetime"] = pd.to_datetime(
    df["start_datetime"],
    errors="coerce"
)


df["end_datetime"] = pd.to_datetime(
    df["end_datetime"],
    errors="coerce"
)


df["duration_minutes"] = (
    df["end_datetime"]
    -
    df["start_datetime"]
).dt.total_seconds()/60

df[
[
"event_cause",
"duration_minutes"
]
].head()
```

```python
hotspots = (
    df[df.is_obstruction_event==1]
    .groupby(
        [
            "junction",
            "zone"
        ]
    )
    .agg(
        obstruction_count=("id","count"),

        avg_duration=(
            "duration_minutes",
            "mean"
        ),

        trucks=(
            "veh_type",
            lambda x:
            (x=="truck").sum()
        )
    )
    .reset_index()
)
hotspots.sort_values(
    "obstruction_count",
    ascending=False
).head(10)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()


cols=[
    "obstruction_count",
    "avg_duration",
    "trucks"
]


hotspots[
[c+"_scaled" for c in cols]
]=scaler.fit_transform(
    hotspots[cols]
)
hotspots["risk_score"] = (

0.5 *
hotspots["obstruction_count_scaled"]

+

0.3 *
hotspots["avg_duration_scaled"]

+

0.2 *
hotspots["trucks_scaled"]

)*100

hotspots.sort_values(
    "risk_score",
    ascending=False
).head(10)


df["hour"] = (
    df.start_datetime.dt.hour
)


peak_times = (
df[df.is_obstruction_event==1]
.groupby(
[
"junction",
"hour"
]
)
.size()
.reset_index(name="events")
)


peak_times.sort_values(
"events",
ascending=False
).head()
```

```python
risk_df = (
    df[df.is_obstruction_event==1]
    .groupby(
        [
            "junction",
            "zone",
            "hour"
        ]
    )
    .agg(

        event_frequency=(
            "id",
            "count"
        ),

        avg_duration=(
            "duration_minutes",
            "mean"
        ),

        truck_count=(
            "veh_type",
            lambda x:
            x.astype(str)
             .str.lower()
             .str.contains("truck")
             .sum()
        ),

        road_closure_count=(
            "requires_road_closure",
            "sum"
        )
    )
    .reset_index()
)

risk_df.head()
```
